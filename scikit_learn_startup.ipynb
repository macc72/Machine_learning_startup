### importing the necessary libraries

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

### now we can set the variables, models and predict the outcome

```python
x = np.arange(10).reshape(-1, 1)

y = np.array([1, 3, 2, 5, 7, 8, 8, 9, 10, 12])

model = LinearRegression().fit(x, y)

x_new = np.array([[11], [12], [13]])
y_pred = model.predict(x_new)

print(y_pred)
```

### Visualization

```python
x = x = np.arange(10)
plt.plot(x, y)
plt.xticks(x)

plt.show()
```

### try predicting squares of numbers

```python
m  = np.arange(1, 11, 1).reshape(-1, 1)

n = [x**2 for x in m]

model = LinearRegression().fit(m, n)

m_new = np.array([[11], [12], [13]])
n_pred = model.predict(m_new)

plt.plot(m,n)
plt.show()
```

```python
n_new = np.concatenate((n, n_pred))
m_new = np.arange(1, 14, 1)

m_cont = np.arange(10, 14, 1) 
n_cont = [x**2 for x in m_cont]

plt.plot(m_cont, n_cont, 'b:', label= 'correct continuation')

plt.plot(m_new[:10], n_new[:10], label= 'previous trend')
plt.plot(m_new[9:], n_new[9:], 'r:', label='prediction')

plt.xticks(m_new)
plt.legend()
plt.show()
```
